```python
df = pd.read_csv(CSV_FILE)

def get_task_choice(row, task_cols):
    vals = row[task_cols]
    return vals.idxmax()

tasks = {
    "Task01": ["Class1.1", "Class1.2", "Class1.3"],  
    "Task02": ["Class2.1", "Class2.2"],              
    "Task06": ["Class6.1", "Class6.2"],              
    "Task09": ["Class9.1", "Class9.2", "Class9.3"],  
}
```

```python
def traverse_tree(row):
    # Top-level: smooth / features_or_disk / star_or_artifact
    choice01 = get_task_choice(row, tasks["Task01"])

    if choice01 == "Class1.1":  # smooth
        choice06 = get_task_choice(row, tasks["Task06"])
        if choice06 == "Class6.2":   # no odd
            return os.path.join("smooth", "no_odd")
        else:
            return os.path.join("smooth", "odd")

    elif choice01 == "Class1.2":  # features_or_disk
        choice06 = get_task_choice(row, tasks["Task06"])
        if choice06 == "Class6.1":   # yes odd
            return os.path.join("features_or_disk", "odd")

        choice02 = get_task_choice(row, tasks["Task02"])
        if choice02 == "Class2.1":  # edge-on = yes
            choice09 = get_task_choice(row, tasks["Task09"])
            leaf = "no_bulge" if choice09 == "Class9.3" else "bulge"
            return os.path.join("features_or_disk", leaf)
        else:
            return os.path.join("features_or_disk", "no_bulge")

    else:
        return "star_or_artifact"

moved = 0
skipped = 0
```

```python
for _, row in df.iterrows():
    try:
        img_name = str(int(row["GalaxyID"]))
    except Exception:
        img_name = str(row["GalaxyID"]).strip()

    src = os.path.join(IMAGE_DIR, f"{img_name}.jpg")
    if os.path.exists(src):
        dest_folder = os.path.join(OUTPUT_DIR, traverse_tree(row))
        os.makedirs(dest_folder, exist_ok=True)

        dest = os.path.join(dest_folder, f"{img_name}.jpg")
        count = 1
        while os.path.exists(dest):
            dest = os.path.join(dest_folder, f"{img_name}_{count}.jpg")
            count += 1

        shutil.move(src, dest)
        moved += 1
    else:
        skipped += 1
```

```python
def is_image(fname):
    f = fname.lower()
    return f.endswith((".jpg", ".jpeg", ".png"))

def list_leaf_dirs(root):
    root = pathlib.Path(root)
    leaves = []
    for dirpath, _, filenames in os.walk(root):
        files = [f for f in filenames if is_image(f)]
        if files:
            p = pathlib.Path(dirpath)
            # skip root itself if it contains images
            if p == root:
                continue
            rel = p.relative_to(root).parts  # e.g. ('smooth','no_odd') or ('star_or_artifact',)
            class_key = "_".join(rel)        # e.g. 'smooth_no_odd'
            leaves.append((p, class_key))
    # dedupe
    seen, out = set(), []
    for p, k in leaves:
        if k not in seen:
            seen.add(k)
            out.append((p, k))
    return out

leaf_dirs = list_leaf_dirs(LEAF_ROOT)
```

```python
if os.path.exists(DATASET_DIR):
    shutil.rmtree(DATASET_DIR)
os.makedirs(DATASET_DIR, exist_ok=True)

for src_dir, class_key in leaf_dirs:
    dst_dir = pathlib.Path(DATASET_DIR, class_key)
    dst_dir.mkdir(parents=True, exist_ok=True)
    for name in os.listdir(src_dir):
        if not is_image(name): 
            continue
        src = pathlib.Path(src_dir, name)
        dst = pathlib.Path(dst_dir, name)
        if dst.exists():
            # avoid collisions by suffixing a counter
            stem, ext = os.path.splitext(name)
            c = 1
            while dst.exists():
                dst = dst_dir / f"{stem}_{c}{ext}"
                c += 1
        try:
            # Hardlink to avoid duplicates (fast, no extra space)
            os.link(src, dst)
        except Exception:
            # Fallback to copy if hardlinks not permitted
            shutil.copy2(src, dst)
```
