```python
def traverse_tree(row):
    choice01 = get_task_choice(row, tasks["Task01"])
    if choice01 == "Class1.1":  # smooth
        choice06 = get_task_choice(row, tasks["Task06"])
        if choice06 == "Class6.2":  # no odd
            return os.path.join("smooth", "odd_no")
        else:  # yes odd
            choice08 = get_task_choice(row, tasks["Task08"])
            mapping = {
                "Class8.1": "odd_ring",
                "Class8.2": "odd_lens_or_arc",
                "Class8.3": "odd_disturbed",
                "Class8.4": "odd_irregular",
                "Class8.5": "odd_merger",
                "Class8.6": "odd_dust_lane",
                "Class8.7": "odd_other"
            }
            return os.path.join("smooth", mapping.get(choice08, "odd_other"))

    elif choice01 == "Class1.2":  # features_or_disk
        choice02 = get_task_choice(row, tasks["Task02"])
        if choice02 == "Class2.1":  # yes edge-on
            choice09 = get_task_choice(row, tasks["Task09"])
            mapping = {
                "Class9.1": "bulge_rounded",
                "Class9.2": "bulge_boxy",
                "Class9.3": "bulge_no_bulge"
            }
            return os.path.join("features_or_disk", mapping.get(choice09, "bulge_no_bulge"))
        else:  # no edge-on
            choice06 = get_task_choice(row, tasks["Task06"])
            if choice06 == "Class6.2":  # no odd
                return os.path.join("features_or_disk", "bulge_no_bulge")
            else:
                choice08 = get_task_choice(row, tasks["Task08"])
                mapping = {
                    "Class8.1": "odd_ring",
                    "Class8.2": "odd_lens_or_arc",
                    "Class8.3": "odd_disturbed",
                    "Class8.4": "odd_irregular",
                    "Class8.5": "odd_merger",
                    "Class8.6": "odd_dust_lane",
                    "Class8.7": "odd_other"
                }
                return os.path.join("features_or_disk", mapping.get(choice08, "odd_other"))
    else:
        return "star_or_artifact"
```

```python
def is_image(fname):
    f = fname.lower()
    return f.endswith((".jpg", ".jpeg", ".png"))

def list_leaf_dirs(root):
    root = pathlib.Path(root)
    leaves = []
    for dirpath, _, filenames in os.walk(root):
        files = [f for f in filenames if is_image(f)]
        if files:
            p = pathlib.Path(dirpath)
            # skip root itself if it contains images
            if p == root:
                continue
            rel = p.relative_to(root).parts  # e.g. ('smooth','no_odd') or ('star_or_artifact',)
            class_key = "_".join(rel)        # e.g. 'smooth_no_odd'
            leaves.append((p, class_key))
    # dedupe
    seen, out = set(), []
    for p, k in leaves:
        if k not in seen:
            seen.add(k)
            out.append((p, k))
    return out

leaf_dirs = list_leaf_dirs(LEAF_ROOT)
```

```python
if os.path.exists(DATASET_DIR):
    shutil.rmtree(DATASET_DIR)
os.makedirs(DATASET_DIR, exist_ok=True)

for src_dir, class_key in leaf_dirs:
    dst_dir = pathlib.Path(DATASET_DIR, class_key)
    dst_dir.mkdir(parents=True, exist_ok=True)
    for name in os.listdir(src_dir):
        if not is_image(name): 
            continue
        src = pathlib.Path(src_dir, name)
        dst = pathlib.Path(dst_dir, name)
        if dst.exists():
            # avoid collisions by suffixing a counter
            stem, ext = os.path.splitext(name)
            c = 1
            while dst.exists():
                dst = dst_dir / f"{stem}_{c}{ext}"
                c += 1
        try:
            # Hardlink to avoid duplicates (fast, no extra space)
            os.link(src, dst)
        except Exception:
            # Fallback to copy if hardlinks not permitted
            shutil.copy2(src, dst)
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout, RandomFlip, RandomRotation
from tensorflow.keras.models import Model
```

```python
data_augmentation = tf.keras.Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.1),
])

base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(126, 126, 3))
base_model.trainable = False  # Freeze initially

# 2. Build the Model Functional Style
inputs = tf.keras.Input(shape=(126, 126, 3))
x = data_augmentation(inputs)
x = preprocess_input(x)  # Ensure ResNet50 gets the pixel values it expects
x = base_model(x, training=False) # Important: Keep BN layers in inference mode
x = GlobalAveragePooling2D()(x)
x = Dropout(0.5)(x)      # 3. Add Dropout to prevent memorization
predictions = Dense(len(class_names), activation='softmax')(x)

model = Model(inputs, predictions)

# --- Phase 1: Training the Top Layers ---
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Starting Initial Training...")
history = model.fit(train_ds, validation_data=val_ds, epochs=15)

# --- Phase 2: Controlled Fine-Tuning ---
# Instead of unfreezing EVERYTHING, we unfreeze the top layers of ResNet
base_model.trainable = True
for layer in base_model.layers[:-30]: # Keep the first 140+ layers frozen
    layer.trainable = False

# Use a significantly smaller learning rate for fine-tuning
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print("Starting Fine-Tuning...")
# 4. Add EarlyStopping to catch the model before it overfits
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

history_fine = model.fit(train_ds, 
                         validation_data=val_ds, 
                         epochs=20, 
                         callbacks=[early_stop])

# Final Evaluation
final_val_loss, final_val_acc = model.evaluate(val_ds, verbose=0)
print(f"Improved Validation Accuracy: {final_val_acc * 100:.2f}%")
```

```python
def predict_image(image_path, threshold=0.0,
                  model_path="galaxy_classifier.keras",
                  classmap_path="class_names.json"):
    """
    Returns (predicted_class, confidence) and the top-3 list.
    If threshold > 0 and confidence < threshold, returns ('uncertain', confidence).
    """
    mdl = tf.keras.models.load_model(model_path)
    with open(classmap_path, "r") as f:
        classes = json.load(f)

    img = load_img(image_path, target_size=IMG_SIZE)
    arr = img_to_array(img)
    arr = np.expand_dims(arr, 0)  # [1, H, W, 3]
    #arr = arr / 255.0

    probs = mdl.predict(arr, verbose=0)[0]
    top_idx = np.argsort(probs)[::-1]
    top3 = [(classes[i], float(probs[i])) for i in top_idx[:3]]
    pred_idx = top_idx[0]
    pred_class = classes[pred_idx]
    conf = float(probs[pred_idx])

    if threshold and conf < threshold:
        return ("uncertain", conf), top3
    return (pred_class, conf), top3
```

```python
import os
import random
import shutil

def random_image_selection(source_folder: str, num_images: int, new_folder_name: str):
    """
    Selects a random assortment of images from a folder and places them into a new folder.

    Args:
        source_folder (str): Path to the folder containing the images.
        num_images (int): Number of images to randomly select.
        new_folder_name (str): Name of the new folder to store selected images.
    """
    # Get all files in the source folder
    all_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]

    if not all_files:
        print("No files found in the source folder.")
        return

    # Randomly sample images
    selected_files = random.sample(all_files, min(num_images, len(all_files)))

    # Path for the new folder
    new_folder_path = os.path.join(source_folder, new_folder_name)

    # Create the new folder if it doesn't exist
    os.makedirs(new_folder_path, exist_ok=True)

    # Copy selected files into the new folder
    for file_name in selected_files:
        src = os.path.join(source_folder, file_name)
        dst = os.path.join(new_folder_path, file_name)
        shutil.copy(src, dst)

    print(f"Copied {len(selected_files)} images to '{new_folder_path}'.")

# Example usage

```
